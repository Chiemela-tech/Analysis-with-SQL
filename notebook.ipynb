```python
-- Select the product details, quantity, and location of bicycles in stock
SELECT
	product_name,
	category_name,
	list_price,
	quantity,
	store_id
FROM products
	INNER JOIN stocks ON products.product_id = stocks.product_id
	INNER JOIN categories ON products.category_id = categories.category_id
ORDER BY product_name
```

```python
-- Total sales by category
SELECT 
    c.category_name,
    SUM(oi.quantity * (oi.list_price - oi.discount)) AS total_sales
FROM sales.order_items oi
JOIN production.products p ON oi.product_id = p.product_id
JOIN production.categories c ON p.category_id = c.category_id
GROUP BY c.category_name
ORDER BY total_sales DESC;
```

The Mountain Bikes show that they have the highest sales among others. 
The Children Bicycles have the lowest sales. This could be due to several parents not placing priority on children utilizing bicycles. Some parents also think it is highly risky for children to use bicycles.

```python
-- Sales Trend Over Time (Monthly)
SELECT 
    YEAR(o.order_date) AS year,
    MONTH(o.order_date) AS month,
    SUM(oi.quantity * (oi.list_price - oi.discount)) AS total_sales
FROM sales.orders o
JOIN sales.order_items oi ON o.order_id = oi.order_id
GROUP BY YEAR(o.order_date), MONTH(o.order_date)
ORDER BY year, month;
```

```python
-- Sales Trend Over Time (Monthly)
SELECT 
    YEAR(o.order_date) AS year,
    MONTH(o.order_date) AS month,
    SUM(oi.quantity * (oi.list_price - oi.discount)) AS total_sales
FROM sales.orders o
JOIN sales.order_items oi ON o.order_id = oi.order_id
GROUP BY YEAR(o.order_date), MONTH(o.order_date)
ORDER BY year, month;
```

The month of April shows that this was the month where bicycles were sold the most. This could be because of the festive seasons worldwide where people celebrate Easter, and they could think of buying special gifts for themselves or friends and loved ones, including bicycles.

```python
-- Top 5 Best-Selling Products
SELECT TOP 5
    p.product_name,
    SUM(oi.quantity) AS total_quantity_sold
FROM sales.order_items oi
JOIN production.products p ON oi.product_id = p.product_id
GROUP BY p.product_name
ORDER BY total_quantity_sold DESC;
```

```python
-- Top 5 Best-Selling Products
SELECT TOP 5
    p.product_name,
    SUM(oi.quantity) AS total_quantity_sold
FROM sales.order_items oi
JOIN production.products p ON oi.product_id = p.product_id
GROUP BY p.product_name
ORDER BY total_quantity_sold DESC;
```

```python
-- Customer Segmentation Based on Revenue Generated
SELECT 
    c.customer_id,
    CONCAT(c.first_name, ' ', c.last_name) AS customer_name,
    SUM(oi.quantity * (oi.list_price - oi.discount)) AS total_spent
FROM sales.customers c
JOIN sales.orders o ON c.customer_id = o.customer_id
JOIN sales.order_items oi ON o.order_id = oi.order_id
GROUP BY c.customer_id, CONCAT(c.first_name, ' ', c.last_name)
ORDER BY total_spent DESC;
```

```python
-- Inventory Status Across Stores
SELECT 
    s.store_name,
    p.product_name,
    st.quantity AS stock_quantity
FROM production.stocks st
JOIN sales.stores s ON st.store_id = s.store_id
JOIN production.products p ON st.product_id = p.product_id
ORDER BY s.store_name, p.product_name;
```

```python
-- Sales Performance of Staff
SELECT 
    CONCAT(s.first_name, ' ', s.last_name) AS staff_name,
    COUNT(o.order_id) AS total_orders_handled,
    SUM(oi.quantity * (oi.list_price - oi.discount)) AS total_sales_generated
FROM sales.staffs s
JOIN sales.orders o ON s.staff_id = o.staff_id
JOIN sales.order_items oi ON o.order_id = oi.order_id
GROUP BY CONCAT(s.first_name, ' ', s.last_name)
ORDER BY total_sales_generated DESC;
```
